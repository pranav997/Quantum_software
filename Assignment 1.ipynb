## Playing around with qiskit composer


```python
import numpy as np
import qiskit
import matplotlib.pyplot as plt
```

```python
circu=qiskit.QuantumCircuit(2,2)
circu.x(0)
circu.h(0)
circu.cx(0, 1)
circu.measure(range(2),range(2))
circu.draw('mpl')
```

```python
backend_sim=Aer.get_backend('qasm_simulator')
from qiskit.compiler import transpile
job_sim = backend_sim.run(transpile(circu, backend_sim), shots=1024)

# Grab the results from the job.
result_sim = job_sim.result()
counts = result_sim.get_counts(circu)
print(counts)
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

```python
circ=qiskit.QuantumCircuit(3,3)
circ.h(0)
circ.cx(0, 1)
circ.cx(0, 2)
circ.draw('mpl')
```

```python
from qiskit import Aer
backend=Aer.get_backend('statevector_simulator')
job=backend.run(circ)
result=job.result()
output=result.get_statevector(circ)
print(output)
```
